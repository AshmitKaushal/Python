# Loop 

```python
# In Python, a loop is a programming construct used to execute a block of code repeatedly based on a condition. 
# Loops are essential for performing repetitive tasks efficiently. 
# Python provides two types of loops:
# 1.for loop
# 2.while loop
```

```python
# for loop 
# It is used to iterate over the iterable object.
# Or perform a particular task repeatively based on the condition.
# range function is used in loop to generate a sequence of numbers.

# Range function
# The range function in Python is used to generate a sequence of numbers. 
# It is commonly used with loops to iterate over a specific range of values. 
# The numbers generated by range are not stored in memory; instead, it generates numbers on the fly, making it memory efficient.

# Syntax:
# range(start,stop,step)

# start (optional): The starting number of the sequence (inclusive). Defaults to 0 if not provided.
# stop (required): The number at which the sequence stops (exclusive).
# step (optional): The difference between each number in the sequence. Defaults to 1.
```

```python
# It will return the sequences of number from 0 to 9 bcz 10 is exclusive one which is bydefault a stop parameter.
# If only one parameter is passed inside the range function then it is a stop parameter and start parameter is bydefault is 0.
for i in range(10):
    print(i,end=' ')
```

```python
# for loop with the starting value 
# start parameter specify from which number we want the sequence of numbers
for i in range(1,10):
    print(i,end=' ')
```

```python
# for loop with step parameter 
# It is used for the incrementation or decremenation between the two numbers in the sequence
# bydefault is 1.
for i in range(1,11,1):
    print(i,end=' ')
```

```python
# Now the difference between the two number is 2.
# every time i is incremented by 2.
for i in range(1,11,2):
    print(i,end=' ')
```

```python
# Reverse for Loop 
# If we use stop as negative and the start value is greater than stop value then it is reverse loop.

for i in range(10,-1,-1):
    print(i,end=' ')
```

```python
# Now the i value is every time is decremented by 2
for i in range(10,-1,-2):
    print(i,end=' ')
```

```python
# Nested for loop
# A loop inside the another loop is know as nested for loop
# In which for one value of the outer loop innner loop runs completely.

print('----------Matrix----------')
for i in range(1,4):
    for j in range(1,4):
        print(f"{i}{j}",end=' ')
    print()
```

```python
print("Rectangular Star Pattern")
print()
for i in range(1,6):
    for j in range(1,6):
        print("*",end='')
    print()
```

```python
print("Right-Angled Triangle Pattern")
print()
for i in range(1,6):
    for j in range(1,i+1):
        print("*",end='')
    print()
```

```python
print("Right-Angled Number Pyramid")
print()

for i in range(1,6):
    for j in range(1,i+1):
        print(j,end='')
    print()
```

```python
print("Right-Angled Number Pyramid - II")
print()

for i in range(1,6):
    for j in range(1,i+1):
        print(i,end='')
    print()
```

```python
print("Inverted Right Pyramid")
print()

for i in range(5,0,-1):
    for j in range(1,i+1):
        print("*",end='')
    print()
```

```python
print("Star Pyramid")
print()

for i in range(1,6):
    for k in range(5,i,-1):
        print(" ",end='')
    for j in range(1,(2*i-1)+1):
        print("*",end='')
    print()
```

```python
print("Inverted Star Pyramid")
print()

for i in range(5,-1,-1):
    for k in range(5,i,-1):
        print(" ",end='')
    for j in range(1,(2*i+1)+1):
        print("*",end='')
    print()
```

# Iterable and Iterators

```python
'''
"
In Python, an iterator is an object that implements the iterator protocol, which consists of two main methods: __iter__() and __next__(). 
Iterators are used to iterate over sequences of data, allowing you to access elements one at a time without the need to load the entire sequence into memory at once.

Here are the key characteristics and concepts related to iterators in Python:

Iterable:
An iterable is any object in Python that can be iterated over, such as lists, tuples, strings, dictionaries, sets, and more. 
Iterables have an __iter__() method that returns an iterator.

Iterator:
An iterator is an object returned by the __iter__() method of an iterable. 
It implements the __next__() method, which allows you to fetch the next element in the sequence.
When you call __next__() on an iterator, it returns the next item in the sequence until there are no more items, at which point it raises a StopIteration exception.

Iteration Protocol:
The iteration protocol in Python defines how iteration works.
When you use a for loop or an iterator in Python, it follows this protocol:

The iter() function is called on the iterable to obtain an iterator.
The __next__() method is called on the iterator to fetch elements one by one.
When there are no more elements, StopIteration is raised to signal the end of iteration.

"


'''
```

# Why Converting Iterable to Iterator Is Useful

```python
# When you call iter() on an iterable, it returns an iterator, which generates values one at a time without creating a new collection. 
# This is especially useful for large datasets or infinite sequences.

l1 = [1,2,3]
it = iter(l1)
print(next(it))
print(next(it))
```

# But once the list is stored in the memory which has thousands of element , so much memory is already consumed then why there is need of converting iterable into iterator ?


```python
'''
You're correct that if you have a list with thousands of elements, it will already consume memory since all elements are stored in memory at once. 
The need to convert an iterable into an iterator arises not from saving memory in such cases, but rather from enabling certain behaviors and use cases where direct access to elements sequentially (without loading or copying the whole data structure) is beneficial.

Let me explain why converting iterables into iterators is still useful:

1. Sequential Access Without Indexing
Lists and tuples allow random access (e.g., l[0], l[1]), but iterators provide sequential access, often useful for algorithms that process elements one-by-one without needing to revisit them.
Iterators simplify looping operations, especially in cases where you don't need to store or process the entire collection at once.
Example:

python
Copy code
nums = [1, 2, 3, 4, 5]
it = iter(nums)  # Convert iterable to iterator
print(next(it))  # Access one element: 1
print(next(it))  # Access next element: 2

2. Unified Interface for All Iterable Objects
Not all iterables are stored in memory like lists or tuples. 
For example:
range is a lazy iterable.
Generators yield elements on demand.
Converting an iterable into an iterator allows code to work uniformly with all iterable types, whether or not they are memory-intensive.
Example of a single interface for different iterables:

python
Copy code
def process(iterable):
    it = iter(iterable)
    for item in it:
        print(item)

process([1, 2, 3])           # Works with lists
process(range(5))            # Works with lazy ranges
process(x for x in range(3)) # Works with generators

3. Memory-Efficient Chaining or Pipelining
When working with large data or streams, converting an iterable to an iterator ensures that operations happen one step at a time, rather than creating new, potentially massive, intermediate collections.
Example of chaining with map and filter:

python
Copy code
nums = range(1, 1000000)  # Does not store all numbers
result = map(lambda x: x * 2, nums)  # Iterator that doubles numbers
final_result = filter(lambda x: x % 3 == 0, result)  # Iterator that filters divisible by 3

for num in final_result:
    print(num)  # Processes numbers one-by-one

4. Infinite or Streaming Data
Iterators are essential for handling infinite sequences or data streams. Lists cannot handle infinite data since they require all elements to be in memory.
Example:

python
Copy code
import itertools
infinite = itertools.count()  # Infinite iterator starting from 0

for num in infinite:
    if num > 10:
        break
    print(num)

5. Avoiding Copies
Iterators avoid creating unnecessary copies of data. If you loop over an iterator, you don't create a new collection; you process the elements directly.


Conclusion
The primary reasons to convert an iterable into an iterator are functional, not necessarily to save memory for pre-stored data like a list:

Enable sequential, one-by-one access.
Work uniformly with all iterable types (including lazy ones).
Avoid creating copies of data.
Handle infinite or large streams of data.
When dealing with a list, the list already occupies memory, but converting it into an iterator allows on-the-fly processing and integration with other lazy pipelines like map, filter, and zip.
'''
```

# why the list and tuple does not act like range and generator ?

```python
# The reason lists and tuples don't act like range or generators is due to their design purpose and implementation. 
# Each of these data structures or objects serves a specific role, and their behavior is tailored to fulfill those roles efficiently. 
# Here's why lists and tuples are different from range and generators:

```

```python
# 1. Lists and Tuples: Eager Evaluation
# Definition: Lists and tuples are designed to store all elements explicitly in memory.
# Behavior:
# When a list or tuple is created, all elements are calculated and stored in memory.
# You can access elements directly by their index, making them suitable for use cases requiring random access.
# Trade-off: Because all elements are stored in memory, they consume a fixed amount of space proportional to the size of the collection.

l1 = [10,20,30,40]
print(l1[2])
```

```python
# 2. Range: Lazy Evaluation
# Definition: A range object generates numbers lazily, only producing each number on demand.
# Why:
# A range object is meant to represent a sequence of numbers efficiently without consuming memory for all elements.
# It calculates values as needed, which allows it to handle very large sequences.
# Use Case:
# Efficiently loop through a sequence of numbers without storing them.
# Example:

r = range(1,10000000000000)
print(r)
print(type(r))

# Accessing the random element,it is generating the data on demand.
print(r[5])
```

```python
# 3. Generators: Fully Lazy
# Definition: A generator is a more general-purpose tool that produces values one at a time using yield.
# Why:
# Generators are designed for cases where you don't want to calculate all values at once.
# They can produce infinite sequences or work with data streams without ever running out of memory.
# Trade-off: Unlike range, generators cannot support random access (e.g., g[2] is invalid).

def gen():
    for i in range(5):
        yield i
g1 = gen()
print(next(g1))
print(next(g1))
```

```python
# 4. Why Lists and Tuples Don’t Behave Lazily
# Purpose: Lists and tuples are optimized for random access and repeated use.
# A list or tuple allows you to access elements at any index (l[5]) without recalculating the previous values.
# They are mutable (for lists) or immutable (for tuples), making them suitable for certain operations where lazy evaluation isn't helpful.
# Use Case:
# Use lists or tuples when you need to store all elements at once and access them multiple times or in random order.
# Trade-offs:

# Lists/Tuples:
# Store all data in memory, which can be inefficient for large datasets.
# Allow indexing, slicing, and repeated use.
# Range/Generators:
# Efficient in memory for large or infinite datasets.
# Lack direct random access.
```

```python
# 5. Design Philosophy
# Lists and tuples are intended to represent concrete collections of data.
# They are fully realized and stored in memory, making them useful for algorithms that require the entire dataset.
# Range and generators focus on performance and memory efficiency, producing values lazily.
```

```python
# 6. Conclusion
# Lists and tuples prioritize random access and repeated use, so they store all their elements in memory. 
# In contrast, range and generators are designed for efficient, lazy generation of elements, making them ideal for processing large or infinite sequences where you don’t need to store all elements at once. The difference lies in their intended use cases and trade-offs between memory and functionality.
```
